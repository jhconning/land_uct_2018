### Problem 1 (in Lab 1)

A village has communal lands. Resident villagers have rights to farm unutilized lands. There are two farmers A and B who each have access to the same production technology. Each unit of output they produce can be sold to traders at a price of $p=1$ per unit.  Land must be cleared before it can be farmed at a cost of $c_l$ per unit. 

Farmer A however has higher status within the village which gives her the right to choose cultivation land first. Village norms establish that she should not use more land than she can profitably farm, but she does have rights of first access and no rent is charged.

Farmer $i$ chooses the amount of land $T_i$ they want to farm to solve:

$$\max_{T_i}  p \cdot F(T_i) - r \cdot T_i $$

where $F(T_i) = A \cdot 2 \sqrt T_i $

With abundant land and in the absence of any type of land rental market we take $r$ to simply equal to the cost of clearing land $c_l$.

a) Solve the farmer's problem to derive an expression for their optimal land demand function $T_i^*(r, p)$:

**answer:** To maximize farmer profits $\Pi(T_i) = p \cdot A \cdot 2   \sqrt T_i - r \cdot T_i$ we use the first-order necessary condition (FOC) for a maximum:

$$\frac{d\Pi}{dT}= \frac{pA}{\sqrt T} - r = 0$$

to solve for optimal land demand as a funciton of $r, P$ and $A$:

$$T_i^*(r,p) = \left ( \frac{pA}{r} \right )^2$$

f) *in this now land-scarce economy, what are the first-order necessary conditions for efficient land allocation? You can find this by setting up the planner's programming problem.  How would land be allocated?  What is the total quantity and value of output?*

To maximize the value of village output a social planner would solve:

$$\max_{T_A, T_B} \ \ P \cdot F(T_A) + P \cdot F(T_A)$$

$$subject \ \ to$$

$$T_A + T_B \le \bar T$$

When land is scarce the constraint binds and we can substitute $T_B = \bar T - T_A$. With this the problem becomes:

$$\max_{T_A, T_B} \ \ P \cdot F(T_A) + P \cdot F(\bar T - T_A)$$

which leads to  first-order necessary conditions:

$$F_T(T_A) =  F_T(\bar T - T_A)  $$

Substituting in the parameters and solving for efficient $T_A^e$

$$\frac{A}{\sqrt T_A} = \frac{A}{\sqrt {\bar T - T_A}} $$

$$\bar T - T_A = T_A$$

$$T_A^e = T_B^e =T^e \frac{\bar T}{2}$$

Since we assumed $\bar T = 100$ we find the efficient allocation is for each to farm $T^e=50$
At this level the shadow rental $r$ is:

$$r = F_T(T^e) = p \cdot \frac{A}{\sqrt T^e}$$

which in and economy with $p=2$ and $A=4$ gives  $r =2 \cdot \frac{4}{\sqrt 50} \approx 1.13$

You can easily verify that if the farmers faced this rental rate they would each demand the efficient level of land that just exhausts the land available.

If each farmer uses $T^e$ they each produce $F(T^e) = 2\cdot A \sqrt \frac{\bar T}{2} =8 \cdot \sqrt{50} \approx 56.6$ for total output of $2 \cdot F(T^e) = 16 \cdot \sqrt{50} \approx 113.14$

At a price of p=\$2 that is close to \$226.28.

Compare this to the situation where A gets as much land as they could clear and B the remainder. Total output was 112 which at p=\$2 is \$224 as the value of total output. 

g) *What is the deadweight loss (DWL) associated with the allocation where A gets their preferred land allocation and B only gets what remains?* (hint: how much lower is the value of output compared to an efficient allocation).

*Does this DWL seem large or small  In practice what factors not considered here might make the DWL higher or lower?*

**answer**: The value of output with efficient allocation is \$226.28 compared to \$224 in the absence of an efficient land allocation mechanism.  The deadweight loss (DWL) or cost of the misallocation is therefore measured as the difference \$226.28 - \$224 $\approx \$2.28 $

You may be surprised that this does not seem to be a great deal of money, only about 1/2 of one percent of the value of output under efficient allocation. 

More generally the formula for DWL here would be:

$$DWL(T_A) =  2 \cdot A \cdot [N \cdot \sqrt{\bar T/N} - (N-1)\cdot(\sqrt{(\bar T-T_a)}) ] $$

When $N=2$ we get the same results as described above. When $N>2$ the interpretation is that farmer A gets their preferred claim and all the other farmers then divide up the remaining lands equally. Given that the technology displays diminishing marginal product of land, adding farmers and their farm technology can raise farm output and productivity putting more pressure on land and making farmer A's large allocation more wasteful.

In other contexts DWL costs might be higher, and it should also be remembered that this would be a recurring cost if repeated year after year. Other things that might make the costs higher: (1) Farmer B might be a more skilled farmer (e.g. his A parameter is higher) and this would increase the cost of over-allocating land to the less skilled farmer A. (2) If the cost of clearing is lower then both farmers want more land but if we continue to grant it to A the DWL increases, (3) Higher A or P or more farmers would increase DWL by increasing the value productivity of land.  In this last sense it might be that DWL is low in low-productivity agriculture, and hence the benefits of 'reforms' to create a land market perhaps not large enough to justify the costs.

A lesson learned here is that it's worth investigating the empirical costs of missing land markets and property rights before embarking on potentially expensive and disruptive reforms.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, fixed
%matplotlib inline
```

```python
A = 4
p = 1
cl = 1
Tbar = 100
```

```python
def F(T, A=A):
    return 2*A*np.sqrt(T)

def MVPT(p,T,A=A):
    '''marginal value product of labor P*MPT(T) or land demand curve'''
    return A*p/T**(1/2)

def TD(p,r,A=A):
    '''Land demand as a function of (r,p)'''
    return (p*A/r)**2
```

It will be useful to know the shadow rental price of land (or competitive land rental rate).  This can be calculated from:

$$F_T(T_A) = F_T(\bar T - T_A)$$

$$\frac{A}{\sqrt T_A} = \frac{A}{\sqrt {\bar T - T_A}} $$

$$\bar T - T_A = T_A$$

$$T_A^e = T_B^e =T^e \frac{\bar T}{2}$$

Since we assumed $\bar T = 100$ we find the efficient allocation is for each to farm $T^e=50$
At this level the shadow rental $r$ is:

$$r = F_T(T^e) = p \cdot \frac{A}{\sqrt T^e}$$

which in and economy with $p=2$ and $A=4$ gives  $r =2 \cdot \frac{4}{\sqrt 50} \approx 1.13$

```python
def TE(P,A=A, Tb = Tbar, N=2):
    '''efficient land allocation'''
    return Tb/N

def efft(P, cl, Tb=Tbar, N=2, A=A):
    '''efficient land use and (shadow) rental rate:
       will return r=cl if total demand for land is less than 
       economy endowment Tb'''
    td = TD(P,cl,A)
    te = TE(P,A,Tb,N)
    if N*td < Tb:
        return td, cl
    else:
        return te, P*A/np.sqrt(TE(P,A,Tb,N))
```

```python
td, re = efft(P=2,cl=1,Tb=100,N=2,A=4)
td, re
```

```python
TD(2,cl,A)
```

```python
t = np.linspace(0.1,Tbar-0.1, 2*Tbar)
def landmarket(P, cl, title, A=A):
    '''land allocation plots'''
    fig = plt.subplots(figsize=(8,6))
    td, re = efft(P,cl,Tb=100,N=2,A=4)
    ta = TD(P,cl)
    plt.ylim(0,5)
    plt.axhline(max(cl,re),linestyle='--')
    plt.axhline(cl,linestyle=':')
    plt.plot(t,MVPT(P,t))
    plt.text(0.7*Tbar, MVPT(P,0.7*Tbar),r'$P \cdot F_T(T_A)$', fontsize=18)
    plt.text(0.1*Tbar, MVPT(P,Tbar*0.7),r'$P \cdot F_T(\bar T - T_A)$', fontsize=18)
    plt.xlabel('T -- land use', fontsize=18)
    plt.ylabel('MVPT', fontsize=18)
    plt.scatter(td, MVPT(P,td))
    plt.scatter(Tbar-td,MVPT(P,td))
    plt.plot([td,td],[0,MVPT(P,td)],':')
    plt.plot([ta,ta],[0,MVPT(P,Tbar-ta)],':')  #T_A
    plt.plot([Tbar-td,Tbar-td],[0,MVPT(P,td)],':')
    plt.plot(t,MVPT(P,Tbar - t))
    plt.fill_between(t,MVPT(P,Tbar - t),MVPT(P,t),
                     where =(MVPT(P,Tbar - t)>MVPT(P,t))*(ta>td)*(t<ta))
    plt.title(title)
    plt.xlim(0,Tbar);
```

```python
def DWL(Ta, A, P, N):
    qe = N*P*2*A*np.sqrt(Tbar/N)
    return qe, (qe - P*2*A*(np.sqrt(Ta)+(N-1)*np.sqrt((Tbar-Ta)/(N-1))))/qe
```
